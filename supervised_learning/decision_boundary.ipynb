```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])

x_s = []
x_c = []
for i in range(len(X)):
    if y[i] == 1:
        x_s.append(X[i, :])
    else:
        x_c.append(X[i, :])   
        
x_s = np.array(x_s)
x_c = np.array(x_c)
```

```python
plt.scatter(x_c[:, 0], x_c[:, 1], color="black", s=100, marker="o")
plt.scatter(x_s[:, 0], x_s[:, 1], color="red", s=100, marker="X")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
z = np.arange(-10, 11)
s = 1 / (1 + np.exp(-z))

plt.plot(z, s)
plt.xlabel("Z")
plt.ylabel("S")
plt.title("Sigmoid Function")
plt.show()
```

```python
plt.scatter(x_c[:, 0], x_c[:, 1], color="black", s=100, marker="o")
plt.scatter(x_s[:, 0], x_s[:, 1], color="gray", s=100, marker="X")
plt.plot(np.arange(.3, 3.2), np.arange(.5, 3)*-.8 + 2.50, color="red", linewidth=3)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Decision Boundary")
plt.show()
```
