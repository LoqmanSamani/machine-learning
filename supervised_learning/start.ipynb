<h4>Model Parameters</h4>

```python
w = 100
b = 100
```

```python
def func(x_train, w, b):
    f_wb = w * x_train + b
    return f_wb
```

```python
y_hat = func(x_train, w, b)

plt.plot(x_train, y_hat, c="r", label="Prediction Line")
plt.scatter(x_train, y_train, label="Actual Values")
plt.title("Actual vs. Predicted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

If we solve the system of equation below:
- w * 1 + b = 300
- w * 2 + b = 500

then:
 - w = 200
 - b = 100

```python
y_hat = func(x_train=x_train, w=200, b=100)

plt.plot(x_train, y_hat, c="r", label="Prediction Line")
plt.scatter(x_train, y_train, label="Actual Values")
plt.title("Actual vs. Predicted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Generate some data based on the func()

```python
def func2(func, m, w, b):
    x = np.linspace(1, 2, m)
    y = func(x, w, b)
    sigma = np.array([np.random.randn()*20 for _ in range(m)])
    y_hat = y + sigma
    
    return x, y_hat
```

```python
x_1, y_hat1 = func2(func, 100, 200, 100)

plt.plot(np.array([1.0, 2.0]), np.array([300.0, 500.0]), c="r", label="Prediction Line")
plt.scatter(x_1, y_hat1, label="Actual Values")
plt.title("Actual vs. Predicted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
