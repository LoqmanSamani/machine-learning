```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def compute_cost(x, y, w, b):
    y_hat = w * x + b
    cost = np.mean(np.sum((y_hat - y)**2)) / 2
    return cost
```

```python
np.random.seed(43)
w = 200
b = 100
x = np.linspace(0, 10, 500)
y = w * x + b
y1 = np.array([np.random.randn()*400 + y[i] for i in range(len(x))])
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Regression Line", c="red")
plt.scatter(x, y1, label="Synthetic Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best Fit")
plt.legend()
plt.show()
```

```python
ws = np.array(range(175, 226))
bs = np.array(range(75, 126))

j_wb = np.zeros((len(ws), len(bs)))
for i, w in enumerate(ws):
    for j, b in enumerate(bs):
        cost = compute_cost(x, y1, w, b)
        j_wb[i, j] = cost
        
```

```python
W, B = np.meshgrid(ws, bs)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, j_wb, cmap='viridis')

ax.set_xlabel('W')
ax.set_ylabel('B')
ax.set_zlabel('Cost')

plt.show()
```

```python
W, B = np.meshgrid(ws, bs)
fig, ax = plt.subplots(figsize=(10, 8))
contour = ax.contour(W, B, j_wb, levels=20, cmap='coolwarm')

# Overlay scatter points at specific levels
levels_to_plot = [20, 30, 40, 50, 60, 70, 80, 100, 150, 200]
ax.scatter([], [], c='black', label='Cost Levels')
for level in levels_to_plot:
    ax.scatter([], [], c='black', label=f'Cost {level}', alpha=0.5)

# Add legend
ax.legend()

ax.set_xlabel('W')
ax.set_ylabel('B')

plt.show()
```
