<h3>Batch Gradient Descent Algorithm</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

<h4>Implement squared error cost function</h4>

```python
def compute_cost(y_train, y_hat):
    cost = np.mean(np.power(y_hat - y_train, 2)) / 2
    return cost
```

<h4>Implement forward propagation</h4>

```python
def forward_propagation(x_train, w, b):
    y_hat = w * x_train + b
    return y_hat
```

<h4>Implement gradient descent</h4>

```python
def gradient_descent(x_train, y_train, w, b, alpha):
    m = x_train.shape[0]
    y_hat = forward_propagation(x_train, w, b)
    w -= (alpha * np.sum((y_hat - y_train) * x_train)) / m
    b -= (alpha * np.sum(y_hat - y_train)) / m
    return w, b
    
```

<h4>Train the model</h4>

```python
def train(x_train, y_train, w, b, alpha, n):
    cost = []
    ws = []
    bs = []
    for _ in range(n):
        y_hat = forward_propagation(x_train, w, b)
        j = compute_cost(y_train, y_hat)
        w, b = gradient_descent(x_train, y_train, w, b, alpha)
        cost.append(round(j, 2))
        ws.append(round(w, 2))
        bs.append(round(b, 2))
        
    return w, b, cost, ws, bs
```

```python
w, b, cost, ws, bs = train(x_train, y_train, 0, 0, .01, 100)

print(f"W: {w}")
print(f"b: {b}")
print(f"Cost: {cost}")
print(f"Ws: {ws}")
print(f"bs: {bs}")
```

```python
w1, b1, cost1, ws1, bs1 = train(x_train, y_train, 200, 400, .01, 100)
print(f"W1: {w1}")
print(f"b1: {b1}")
print(f"Cost1: {cost1}")
print(f"Ws1: {ws1}")
print(f"bs1: {bs1}")
```
