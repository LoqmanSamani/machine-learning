<h4><b>Rectified Linear Unit (ReLU) Activation Function</b></h4>

         a = max(0, z)    # ReLU Function

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.activations import sigmoid, relu, linear
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 3, 1)
plt.plot(np.linspace(-10, 10, 50), np.linspace(-10, 10, 50), color="red", linewidth=2.5)
plt.plot(np.zeros(50), np.linspace(-10, 10, 50), linewidth=0.8, color="gray")
plt.plot(np.linspace(-10, 10, 50), np.zeros(50), linewidth=0.8, color="gray")
plt.title("Linear Activation Function")

plt.subplot(1, 3, 2)
plt.plot(np.linspace(-10, 10, 50), 1/(1+np.exp(-np.linspace(-10, 10, 50))),
         color="red", linewidth=2.5)
plt.plot(np.zeros(50), np.linspace(0, 1, 50), linewidth=0.8, color="gray")
plt.title("Sigmoid Activation Function")
plt.subplot(1, 3, 3)
plt.plot(np.linspace(-10, 10, 50), np.maximum(0, np.linspace(-10, 10, 50)),
         color="red", linewidth=2.5)
plt.plot(np.zeros(50), np.linspace(0, 10, 50), color="gray", linewidth=.8)
plt.title("ReLU Acrivation Function")
plt.show()
```
