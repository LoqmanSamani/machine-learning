````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import tensorflow as tf
from tensorflow.keras.layers import Dense, Normalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.callbacks import LambdaCallback
from tensorflow.keras.optimizers.experimental import SGD

from sklearn.datasets import make_classification
````

<p> The tf.keras.Input(shape=(2,)), specifies the expected shape of the input. This allows Tensorflow to size the weights and bias parameters at this point. This is useful when exploring Tensorflow models. This statement can be omitted in practice and Tensorflow will size the network parameters when the input data is specified in the model.fit statement.</p>
```python
tf.random.set_seed(1234)  # applied to achieve consistent results
model = Sequential(
    [
        tf.keras.Input(shape=(2,)),
        Dense(3, activation='sigmoid', name = 'layer1'),
        Dense(1, activation='sigmoid', name = 'layer2')
     ]
)
```


````python
X, Y = make_classification(
    n_samples=2000,  
    n_features=5,    
    n_informative=5,  
    n_redundant=0,    
    n_clusters_per_class=1,
    random_state=42
)
print(X[:20])
print(Y[:20])
print(X.shape)
print(Y.shape)
````

````python
plt.scatter( X[:200, 0], X[:200, 1], color="red", label="X1 vs. X2")
plt.scatter(X[:200, 0],  X[:200, 2], color="blue", label="X1 vs. X3")
plt.scatter( X[:200, 0], X[:200, 3], color="green", label="X1 vs. X4")
plt.scatter( X[:200, 0], X[:200, 4], color="green", label="X1 vs. X5")

plt.legend()
plt.show()
````

````python
X_train = X[:1900]
X_test = X[1900:]
Y_train = Y[:1900].reshape((1900, 1))
Y_test = Y[1900:]
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
````

````python
def print_cost(epoch, logs):
    if epoch % 10 == 0:
        print("Epoch {}: Cost: {:.4f}".format(epoch, logs["loss"]))
````

````python

model = Sequential([
    Dense(units=5, activation="relu"),
    Dense(units=4, activation="relu"),
    Dense(units=2, activation="relu"),
    Dense(units=1, activation="sigmoid")
])

model.compile(optimizer=SGD(learning_rate=0.01), loss=BinaryCrossentropy())
callback = LambdaCallback(on_epoch_end=print_cost)

model.fit(X_train, Y_train, epochs=200, verbose=0, callbacks=[callback])
````

````python
predict = model.predict(X_test)
print(predict[:20])
````

````python
pred = (predict >= 0.5).astype(int)
print(pred[:20])
````

````python
pred = pred.reshape((1, 100)).flatten()
accuracy = np.sum([1 if i==j else 0 for i,j in zip(pred, Y_test)])/len(Y_test)
print(accuracy)
````
